```python
def plot_histograms_by_default(data, target_variable='SeriousDlqin2yrs', num_variables=None):
    """
    Genera histogramas múltiples para comparar la distribución de variables entre
    clientes que incumplieron y los que no incumplieron.
    
    Args:
        data: DataFrame con los datos
        target_variable: Nombre de la variable objetivo (default/no default)
        num_variables: Número de variables a graficar (None = todas)
    """
    # Excluimos la variable objetivo y columnas no numéricas
    exclude_cols = [target_variable]
    if 'Unnamed: 0' in data.columns:
        exclude_cols.append('Unnamed: 0')
    
    numeric_cols = [col for col in data.columns if col not in exclude_cols]
    
    # Si se especifica un número de variables, tomamos las primeras n
    if num_variables and num_variables < len(numeric_cols):
        numeric_cols = numeric_cols[:num_variables]
    
    # Creamos una figura con subplots
    n_cols = 2  # número de columnas en la figura
    n_rows = (len(numeric_cols) + 1) // n_cols  # número de filas necesarias
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
    axes = axes.flatten()  # Convertimos a un array 1D para fácil indexación
    
    # Preparamos las etiquetas para la leyenda
    labels = {0: 'No incumplió', 1: 'Incumplió'}
    
    # Colores para cada grupo
    colors = {0: 'blue', 1: 'red'}
    
    # Generamos histogramas para cada variable
    for i, col in enumerate(numeric_cols):
        ax = axes[i]
        
        # Filtramos outliers para visualización
        col_data = data[[col, target_variable]].copy()
        
        # Manejo especial para variables con valores extremos
        if col == 'RevolvingUtilizationOfUnsecuredLines':
            col_data = col_data[col_data[col] < 10]
        elif col == 'DebtRatio':
            percentile = col_data[col].quantile(0.95)
            col_data = col_data[col_data[col] < percentile]
        
        # Grupos por la variable objetivo
        for target_val in [0, 1]:
            subset = col_data[col_data[target_variable] == target_val]
            
            # Normalizamos los histogramas para comparar las distribuciones
            ax.hist(subset[col], bins=30, alpha=0.5, 
                    color=colors[target_val], 
                    label=labels[target_val],
                    density=True)  # Normalizar por densidad
        
        ax.set_title(f'Distribución de {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Densidad')
        ax.legend()
        ax.grid(True, linestyle='--', alpha=0.7)
    
    # Ocultamos los ejes vacíos si los hay
    for j in range(i+1, len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    return fig

def plot_class_distribution(data, target_variable='SeriousDlqin2yrs'):
    """
    Visualiza la distribución de la variable objetivo (incumplimiento).
    """
    # Conteo de valores
    target_counts = data[target_variable].value_counts().sort_index()
    
    # Calcular porcentajes
    total = target_counts.sum()
    percentages = (target_counts / total * 100).round(2)
    
    # Crear etiquetas
    labels = {
        0: f'No incumplió: {target_counts[0]} ({percentages[0]}%)',
        1: f'Incumplió: {target_counts[1]} ({percentages[1]}%)'
    }
    
    # Crear gráfico
    plt.figure(figsize=(10, 6))
    bars = plt.bar([labels[0], labels[1]], target_counts, color=['blue', 'red'])
    
    # Añadir valores en las barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                 f'{height:,}', ha='center', va='bottom')
    
    plt.title('Distribución de Incumplimientos (SeriousDlqin2yrs)')
    plt.ylabel('Número de clientes')
    plt.grid(True, linestyle='--', alpha=0.7, axis='y')
    plt.tight_layout()
    return plt.gcf()
```

```python
# Cargar el dataset
print("Cargando datos...")
data = pd.read_csv("data/cs-training.csv")

# Limpiar el dataset
print("Limpiando datos...")
if "Unnamed: 0" in data.columns:
    data = data.drop(columns=["Unnamed: 0"])

# Eliminar outliers extremos para mejor visualización
data = data[data["RevolvingUtilizationOfUnsecuredLines"] < 13]
data = data[~data["NumberOfTimes90DaysLate"].isin([96, 98])]
data = data[~data["NumberOfTime60-89DaysPastDueNotWorse"].isin([96, 98])]
data = data[~data["NumberOfTime30-59DaysPastDueNotWorse"].isin([96, 98])]
debt_ratio_threshold = data["DebtRatio"].quantile(0.975)
data = data[data["DebtRatio"] <= debt_ratio_threshold]

# Imputar valores faltantes para MonthlyIncome
data["MonthlyIncome"] = data["MonthlyIncome"].fillna(data["MonthlyIncome"].median())

# Mostrar información básica
print("\nInformación del dataset:")
print(f"Total de registros: {len(data)}")
print(f"Distribución de SeriousDlqin2yrs:")
default_counts = data['SeriousDlqin2yrs'].value_counts().sort_index()
for value, count in default_counts.items():
    print(f"  {value} ({'Incumplió' if value == 1 else 'No incumplió'}): {count} ({count/len(data)*100:.2f}%)")
```

```python
# Visualizar distribución de la variable objetivo
print("\nGenerando gráfico de distribución de clases...")
fig_dist = plot_class_distribution(data)
plt.show()
```

```python
# Visualizar histogramas múltiples
print("\nGenerando multi-histogramas por variable...")
fig_hist = plot_histograms_by_default(data)
plt.show()
```
