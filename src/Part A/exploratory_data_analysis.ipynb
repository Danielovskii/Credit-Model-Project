```python
def plot_histograms_by_default(data, target_variable='SeriousDlqin2yrs', num_variables=None):
    """
    Generates multiple histograms to compare the distribution of variables between
    clients who defaulted and those who did not.
    
    Args:
        data: DataFrame with the data
        target_variable: Name of the target variable (default/no default)
        num_variables: Number of variables to plot (None = all)
    """
    # Exclude target variable and non-numeric columns
    exclude_cols = [target_variable]
    if 'Unnamed: 0' in data.columns:
        exclude_cols.append('Unnamed: 0')
    
    numeric_cols = [col for col in data.columns if col not in exclude_cols]
    
    # If a number of variables is specified, select the first n
    if num_variables and num_variables < len(numeric_cols):
        numeric_cols = numeric_cols[:num_variables]
    
    # Create a figure with subplots
    n_cols = 2  # number of columns in the figure
    n_rows = (len(numeric_cols) + 1) // n_cols  # number of rows needed
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
    axes = axes.flatten()  # Flatten to 1D array for easy indexing
    
    # Labels for the legend
    labels = {0: 'Did not default', 1: 'Defaulted'}
    
    # Colors for each group
    colors = {0: 'blue', 1: 'red'}
    
    # Generate histograms for each variable
    for i, col in enumerate(numeric_cols):
        ax = axes[i]
        
        # Filter outliers for better visualization
        col_data = data[[col, target_variable]].copy()
        
        # Special handling for variables with extreme values
        if col == 'RevolvingUtilizationOfUnsecuredLines':
            col_data = col_data[col_data[col] < 10]
        elif col == 'DebtRatio':
            percentile = col_data[col].quantile(0.95)
            col_data = col_data[col_data[col] < percentile]
        
        # Group by target variable
        for target_val in [0, 1]:
            subset = col_data[col_data[target_variable] == target_val]
            
            # Normalize histograms to compare distributions
            ax.hist(subset[col], bins=30, alpha=0.5, 
                    color=colors[target_val], 
                    label=labels[target_val],
                    density=True)  # Normalize by density
        
        ax.set_title(f'Distribution of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Density')
        ax.legend()
        ax.grid(True, linestyle='--', alpha=0.7)
    
    # Hide empty axes if any
    for j in range(i+1, len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    return fig

def plot_class_distribution(data, target_variable='SeriousDlqin2yrs'):
    """
    Visualizes the distribution of the target variable (default).
    """
    # Count values
    target_counts = data[target_variable].value_counts().sort_index()
    
    # Calculate percentages
    total = target_counts.sum()
    percentages = (target_counts / total * 100).round(2)
    
    # Create labels
    labels = {
        0: f'Did not default: {target_counts[0]} ({percentages[0]}%)',
        1: f'Defaulted: {target_counts[1]} ({percentages[1]}%)'
    }
    
    # Create bar chart
    plt.figure(figsize=(10, 6))
    bars = plt.bar([labels[0], labels[1]], target_counts, color=['blue', 'red'])
    
    # Add values on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                 f'{height:,}', ha='center', va='bottom')
    
    plt.title('Default Distribution (SeriousDlqin2yrs)')
    plt.ylabel('Number of clients')
    plt.grid(True, linestyle='--', alpha=0.7, axis='y')
    plt.tight_layout()
    return plt.gcf()
```

```python
# Load the dataset
print("Loading data...")
data = pd.read_csv("data/cs-training.csv")

# Clean the dataset
print("Cleaning data...")
if "Unnamed: 0" in data.columns:
    data = data.drop(columns=["Unnamed: 0"])

# Remove extreme outliers for better visualization
data = data[data["RevolvingUtilizationOfUnsecuredLines"] < 13]
data = data[~data["NumberOfTimes90DaysLate"].isin([96, 98])]
data = data[~data["NumberOfTime60-89DaysPastDueNotWorse"].isin([96, 98])]
data = data[~data["NumberOfTime30-59DaysPastDueNotWorse"].isin([96, 98])]
debt_ratio_threshold = data["DebtRatio"].quantile(0.975)
data = data[data["DebtRatio"] <= debt_ratio_threshold]

# Impute missing values for MonthlyIncome
data["MonthlyIncome"] = data["MonthlyIncome"].fillna(data["MonthlyIncome"].median())

# Show basic information
print("\nDataset information:")
print(f"Total records: {len(data)}")
print(f"Distribution of SeriousDlqin2yrs:")
default_counts = data['SeriousDlqin2yrs'].value_counts().sort_index()
for value, count in default_counts.items():
    print(f"  {value} ({'Defaulted' if value == 1 else 'Did not default'}): {count} ({count/len(data)*100:.2f}%)")
```

```python
# Visualize target variable distribution
print("\nGenerando gráfico de distribución de clases...")
fig_dist = plot_class_distribution(data)
plt.show()
```

```python
# Visualize multiple histograms
print("\nGenerando multi-histogramas por variable...")
fig_hist = plot_histograms_by_default(data)
plt.show()
```
